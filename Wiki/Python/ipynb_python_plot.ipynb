# matplotlib <a class="anchor" id="matplotlib"></a>

the figure (an instance of the class plt.Figure) is a single container that contains all the objects representing axes (ax = plt.axes()), graphics, text, and labels

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = plt.axes() 

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));

# For short, you can use the following codes:
ax.plot(x, x + 4, linestyle='-')  # solid
ax.plot(x, x + 5, linestyle='--') # dashed
ax.plot(x, x + 6, linestyle='-.') # dashdot

ax.plot(x, x + 7, linestyle=':', label='dotted');  # dotted # Labeling Plots
plt.legend(); # add Labeling 

# set labels
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')

# set title
plt.title('Simple Plot')
plt.grid(True)

plt.savefig("pyton_plot_test.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
plt.show()
```

```python
import matplotlib.pyplot as plt

rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
plt.show()
```

```python
import matplotlib.pyplot as plt

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar();  # show color scale
plt.show()
```

## multiple plots

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
```

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```
